```python
import numpy as np
import pandas as pd
import plotly
import plotly.graph_objs as go 
import plotly.express as px
import matplotlib.pyplot as plt 
import schedule
import time
import requests
from io import BytesIO
```

```python
def get_data():
    # Hacer la llamada a la API para el aeropuerto P. Mallorca
    url_mallorca = "https://dataestur.azure-api.net/API-SEGITTUR-v1/AENA_DESTINOS_DL?desde%20%28a%C3%B1o%29=2016&desde%20%28mes%29=01&Aeropuerto%20AENA=P.%20Mallorca"
    r_mallorca = requests.get(url_mallorca)
    df_mallorca = pd.read_excel(BytesIO(r_mallorca.content))

    # Hacer la llamada a la API para el aeropuerto Menorca
    url_menorca = "https://dataestur.azure-api.net/API-SEGITTUR-v1/AENA_DESTINOS_DL?desde%20%28a%C3%B1o%29=2016&desde%20%28mes%29=01&Aeropuerto%20AENA=Menorca"
    r_menorca = requests.get(url_menorca)
    df_menorca = pd.read_excel(BytesIO(r_menorca.content))

    #Llamada a la API para el aeropuerto Ibiza
    url_ibiza = "https://dataestur.azure-api.net/API-SEGITTUR-v1/AENA_DESTINOS_DL?desde%20%28a%C3%B1o%29=2016&desde%20%28mes%29=01&Aeropuerto%20AENA=Ibiza"
    r_ibiza = requests.get(url_ibiza)
    df_ibiza = pd.read_excel(BytesIO(r_ibiza.content))

    #Llamada a la API para el aeropuerto Son Bonet
    url_sonbonet = "https://dataestur.azure-api.net/API-SEGITTUR-v1/AENA_DESTINOS_DL?desde%20%28a%C3%B1o%29=2016&desde%20%28mes%29=01&Aeropuerto%20AENA=Son%20Bonet"
    r_sonbonet = requests.get(url_sonbonet)
    df_sonbonet = pd.read_excel(BytesIO(r_sonbonet.content))

    # Combinar los dos dataframes
    df = pd.concat([df_mallorca, df_menorca, df_ibiza, df_sonbonet])

    #Renombrando columnas
    df = df.rename(columns={'AÑO': 'Año', 'MES': 'Mes', 'AEROPUERTO_AENA': 'Aeropuerto', 'PASAJEROS_POR_DESTINO': 'Pasajeros'})

    # Crear una nueva columna con la fecha completa
    df['fecha'] = pd.to_datetime(df['Año'].astype(str) + "-" + df['Mes'].astype(str), format='%Y-%m')

    # Agrupar por destino y fecha y sumar los pasajeros
    df_agg = df.groupby(['Aeropuerto', 'fecha']).agg({'Pasajeros': 'sum'}).reset_index()
    
    #Visual con Plotly

    fig = px.line(df_agg, x='fecha', y='Pasajeros', color='Aeropuerto', template='none', 
              labels={'Pasajeros': 'Número de pasajeros'}, color_discrete_sequence=['#ff7008', '#474747', '#000000', '#ff6584'])
    fig.update_layout(title='Número de pasajeros por Aeropuerto',
                  plot_bgcolor='rgba(0,0,0,0)')

    fig.show()
```
